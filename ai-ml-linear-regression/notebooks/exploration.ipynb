# Exploratory Data Analysis

This notebook is used for exploratory data analysis (EDA) on the datasets available in the project. It includes visualizations and statistical summaries to understand the data better.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the original sample dataset
sample_data = pd.read_csv('../data/raw/sample_data.csv')

# Load the simulated dataset
simulated_data = pd.read_csv('../data/simulated/simulated_data.csv')

# Display the first few rows of the sample data
sample_data.head()
```

```python
# Display the first few rows of the simulated data
simulated_data.head()
```

```python
# Visualize the relationship between Square Footage and Price in the sample data
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sample_data, x='SquareFootage', y='Price')
plt.title('Square Footage vs Price (Sample Data)')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# Visualize the relationship between Square Footage and Price in the simulated data
plt.figure(figsize=(10, 6))
sns.scatterplot(data=simulated_data, x='SquareFootage', y='Price')
plt.title('Square Footage vs Price (Simulated Data)')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
